```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def find_out(path, num_files):
    df_out = pd.DataFrame()
    for i in range(num_files):
        path_df = path + '/' + str(i) +'_agents_stages_summary_seed_1234.csv'
        df = pd.read_csv(path_df)
        df['Day'] = df.index
        df_out = pd.concat([df_out, df], axis = 0)
    return df_out
```

```python
vacc_cost = 20
test_cost = 5
def find_cumsum_cost(path, num_files):
    df_out = pd.DataFrame()
    for i in range(num_files):
        path_df = path + '/' + str(i) +'_agents_stages_summary_seed_1234.csv'
        df = pd.read_csv(path_df)
        df['Day'] = df.index
        df_out = pd.concat([df_out, df['VACCINE'].cumsum()*vacc_cost + 
                            df['RTPCR_TESTS'].cumsum()*test_cost], axis = 0)
    df_out['Day'] = df_out.index     
    df_out['COST'] = df_out[0]
    return df_out
```

```python
def find_mean_std(path, stage):
    df_out = pd.DataFrame()
    for i in range(10):
        path_df = path + '/' + str(i) +'_agents_stages_summary_seed_1234.csv'
        df = pd.read_csv(path_df)
        df_out = pd.concat([df_out, df[stage]], axis = 1)
    df_mean = df_out.mean(axis = 1)
    df_std = df_out.std(axis = 1)
    return df_mean, df_std
```

```python
def find_peak(path, stage):
    df_out = pd.DataFrame()
    for i in range(10):
        path_df = path + '/' + str(i) +'_agents_stages_summary_seed_1234.csv'
        df = pd.read_csv(path_df)
        df_out = pd.concat([df_out, df[stage]], axis = 1)
    df_max = df_out.max(axis = 0)
    df_mean = df_max.mean()
    df_std = df_max.std()
    return df_mean, df_std   
```

```python
def prob_drop(l1, l2):
    l = []
    for i in range(len(l1)):
        l.append((l1[i]-l2[i])/l1[i])
    return l
```
